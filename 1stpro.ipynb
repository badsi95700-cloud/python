```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time=np.linspace(0,1,1000)
time
```

```python
frequency=5
original_wave=np.sin(2*np.pi*frequency*time)
original_wave
```

```python
frequency=5
counter_wave=np.sin(2*np.pi*frequency*time+np.pi)
counter_wave
```

```python

result=original_wave+counter_wave
result
```

```python
plt.figure(figsize=(10,6))

plt.subplot(3,1,1)
plt.plot(time,original_wave,color="blue")
plt.title("Orginal rf wave")

plt.subplot(3,1,2)
plt.plot(time,counter_wave,color="red")
plt.title("counter rf wave ")

plt.subplot(3,1,3)
plt.plot(time,result,color="green")
plt.title("result after distructive interferance")

plt.tight_layout()
plt.show()

```
